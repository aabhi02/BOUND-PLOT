```python
import random
from matplotlib import pyplot as plt
import pandas as pd
```

```python
rep_num = round(random.uniform(-1, 1), 2)
nums = [round(random.uniform(-1, 1), 2) for _ in range(100)]
nums = [rep_num] * 30 + nums + [rep_num] * 50
df=pd.DataFrame({'numbers':nums})
df.to_excel('bound.xlsx', index=False)
```

```python
plt.axhline(0.75, color='r')
plt.axhline(-0.75, color='r')
plt.plot(nums)
plt.show()
```
